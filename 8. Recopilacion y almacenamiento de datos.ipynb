**weather_conditions: condiciones climáticas en el momento en el que comenzó el viaje**

```python
import pandas as pd
from scipy import stats

# 1️⃣ Cargar el dataset
df = pd.read_csv("/datasets/project_sql_result_07.csv")

# 2️⃣ Convertir start_ts a tipo fecha
df['start_ts'] = pd.to_datetime(df['start_ts'])

# 3️⃣ Revisar las condiciones climáticas únicas
print(df['weather_conditions'].unique())
print(df['weather_conditions'].value_counts())

# 4️⃣ Separar duración de los viajes por grupo usando directamente los valores existentes
duracion_bad = df[df['weather_conditions'] == 'Bad']['duration_seconds']
duracion_good = df[df['weather_conditions'] == 'Good']['duration_seconds']

# 5️⃣ Probar normalidad (opcional)
if len(duracion_bad) >= 3 and len(duracion_good) >= 3:
    print("Normalidad grupo Bad:", stats.shapiro(duracion_bad.sample(500) if len(duracion_bad)>500 else duracion_bad))
    print("Normalidad grupo Good:", stats.shapiro(duracion_good.sample(500) if len(duracion_good)>500 else duracion_good))
else:
    print("No hay suficientes datos para la prueba de normalidad")

# 6️⃣ Prueba t de Student para muestras independientes
t_stat, p_value = stats.ttest_ind(duracion_bad, duracion_good, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, p-value: {p_value:.3f}")

# 7️⃣ Interpretación
alfa = 0.05
if p_value < alfa:
    print("✅ Rechazamos la hipótesis nula: el clima afecta la duración de los viajes.")
else:
    print("❌ No podemos rechazar la hipótesis nula: no hay evidencia de que el clima afecte la duración.")
```

<div class="alert alert-block alert-warning"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Al realizar una prueba de hipótesis, es muy importante verificar los supuestos. El t-test que usaste funciona bien cuando se cumplen ciertos supuestos, como que las varianzas de ambos grupos (sábados lluviosos y no lluviosos) sean similares. Puedes realizar la prueba de levene para comprobar que los dos grupos tienen la misma varianza y, de esta forma, podrás saber el valor del parámetro equal_var que se debe utilizar en la función stats.ttest_ind.

**duration_seconds: duración del viaje en segundos**

```python

import pandas as pd
import numpy as np
from scipy import stats

# 2️⃣ Cargar el dataset
df = pd.read_csv('/datasets/project_sql_result_07.csv')

# 3️⃣ Revisar los valores únicos de weather_conditions
print("Valores únicos de weather_conditions:", df['weather_conditions'].unique())

# 4️⃣ Separar los datos según condiciones climáticas
duracion_bad = df[df['weather_conditions'] == 'Bad']['duration_seconds']
duracion_good = df[df['weather_conditions'] == 'Good']['duration_seconds']

# 5️⃣ Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print("Bad (lluvia/tormenta):")
print(duracion_bad.describe())
print("\nGood (buena):")
print(duracion_good.describe())

# 6️⃣ Probar normalidad (Shapiro-Wilk)
# Si hay más de 500 muestras, se toma un muestreo aleatorio
shapiro_bad = stats.shapiro(duracion_bad.sample(500) if len(duracion_bad) > 500 else duracion_bad)
shapiro_good = stats.shapiro(duracion_good.sample(500) if len(duracion_good) > 500 else duracion_good)

print("\nPrueba de normalidad (Shapiro-Wilk):")
print("Bad:", shapiro_bad)
print("Good:", shapiro_good)

# 7️⃣ Prueba de igualdad de medias (t de Student para muestras independientes)
# Si no cumplen normalidad, se podría usar Mann-Whitney U
t_test_result = stats.ttest_ind(duracion_bad, duracion_good, equal_var=False)  # Welch t-test
print("\nPrueba t de Student (Welch) entre Bad y Good:")
print(t_test_result)

# 8️⃣ Conclusión
alfa = 0.05
if t_test_result.pvalue < alfa:
    print(f"\nSe rechaza la hipótesis nula (p-value={t_test_result.pvalue:.4f}).")
    print("Conclusión: La duración promedio de los viajes cambia los sábados lluviosos.")
else:
    print(f"\nNo se rechaza la hipótesis nula (p-value={t_test_result.pvalue:.4f}).")
    print("Conclusión: No hay evidencia suficiente de que la duración promedio cambie los sábados lluviosos.")
```

**Prueba la hipótesis: La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos.**

```python
import pandas as pd
from scipy import stats

# 1️⃣ Cargar datos
df = pd.read_csv("/datasets/project_sql_result_07.csv")

# 2️⃣ Revisar valores únicos de las condiciones climáticas
print(df['weather_conditions'].unique())

# 3️⃣ Dividir datos por grupo
duracion_bad = df[df['weather_conditions'] == 'Bad']['duration_seconds']
duracion_good = df[df['weather_conditions'] == 'Good']['duration_seconds']

# 4️⃣ Probar normalidad (opcional)
if len(duracion_bad) >= 3:
    print("Normalidad grupo Bad:", stats.shapiro(duracion_bad.sample(min(500, len(duracion_bad)))))
if len(duracion_good) >= 3:
    print("Normalidad grupo Good:", stats.shapiro(duracion_good.sample(min(500, len(duracion_good)))))

# 5️⃣ Probar igualdad de varianzas
levene_stat, levene_p = stats.levene(duracion_bad, duracion_good)
print("Prueba de Levene p-value:", levene_p)

# 6️⃣ Prueba t de Student (igualdad o desigualdad de varianzas según resultado de Levene)
t_stat, p_value = stats.ttest_ind(duracion_bad, duracion_good, equal_var=(levene_p > 0.05))
print("t-statistic:", t_stat)
print("p-value:", p_value)

# 7️⃣ Conclusión
if p_value < 0.05:
    print("Rechazamos H0: La duración promedio cambia los sábados lluviosos")
else:
    print("No rechazamos H0: No hay evidencia de cambio en la duración promedio")
```

**cómo planteaste las hipótesis nula y alternativa**


Para probar si la duración promedio de los viajes cambia en sábados lluviosos, definimos:

Hipótesis nula (H0):
La duración promedio de los viajes desde el Loop hasta O'Hare no cambia en sábados lluviosos.

μBad = μGood


Hipótesis alternativa (H1):
La duración promedio de los viajes desde el Loop hasta O'Hare cambia en sábados lluviosos.

μBad  =μGood

Aquí, Bad corresponde a las horas con lluvia o tormenta, y Good a las demás.


**qué criterio usaste para probar las hipótesis y por qué**

Criterio para probar la hipótesis

Identificación de grupos:

Creamos dos muestras independientes:

duration_bad → duración de viajes en clima malo.

duration_good → duración de viajes en clima bueno.

Verificación de normalidad (opcional):

Usamos la prueba de Shapiro-Wilk para comprobar si cada grupo sigue una distribución aproximadamente normal.

Si las muestras son grandes (>30), el t-test es robusto aunque no sean perfectamente normales.

Comparación de medias:

Se usa el t-test para muestras independientes (Welch t-test) porque los tamaños y varianzas de los dos grupos pueden ser distintos.

El t-test evalúa si la diferencia observada entre las medias es significativa respecto a la variabilidad de los datos.

Decisión:

Si p-value < alfa → se rechaza H0 → la duración promedio cambia.

Si p-value >= alfa → no se rechaza H0 → no hay evidencia suficiente de cambio.
